```python
pd.cut(credit['AGE'],8).head()
labels=['20to32','32to44','44to55','55to67','67to79']
credit["AGE_BIN"] =pd.cut(credit['AGE'],5,labels=labels)
print(pd.value_counts(credit["AGE_BIN"],sort=False))
```

```python
pd.cut(credit['PAY_6'],5).head()
labels=['-2to0','0to2','2to4','4to6','6to8']
credit["PAY_6_BINS"] =pd.cut(credit['PAY_6'],5,labels=labels)
print(pd.value_counts(credit["PAY_6_BINS"],sort=False))
```

```python
pd.cut(credit['BILL_AMT6'],3).head()
labels=['High','Medium','Low']
credit["BILL_AMT_BINS"] =pd.cut(credit['BILL_AMT6'],3,labels=labels)
print(pd.value_counts(credit["BILL_AMT_BINS"],sort=False))
```
